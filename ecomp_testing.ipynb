### Catalog comparison
sel3 was created by merging sel3_origin_filtered.csv and sel3_association_filtered.csv on ORID

sel3 and ugeb bulletins were then trimmed down to only what ORIDs and ARIDs exist in both datasets(note: The ORID column was replaced with EVID for the ugeb dataset, as no values in the ORID column existed in the sel3 dataset)

All timestamps were converted to epoch time, and columns not considered in ecomp comparison were dropped


```python
sel3 = pd.read_csv("sel3_catalog.csv")
sel3.sort_values("ARID")
```

```python
ugeb = pd.read_csv("ugeb_catalog.csv")
ugeb.sort_values("ARID")
```

### Comparison using default thresholds 
Arrival time threshold = 5 seconds

Distance threshold = 250km

```python
ecomp.ecomp('ugeb_catalog.csv', 'sel3_catalog.csv', 'ugeb_sel3_catalog_output.csv')
```

### Comparison using increased thresholds

```python
ecomp.ecomp('ugeb_catalog.csv', 'sel3_catalog.csv', 'ugeb_sel3_catalog_output.csv', arrival_time_tol=50000000, location_tol=500000.0, verbose=True)
```
